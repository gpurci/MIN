Travelling salesman problem
The travelling salesman problem is one of the classic NPhard optimization problems. In this problem, there are n cities
and the distances between the cities are given by a distance
matrix D={dij} (dij the distance between city i and j). There is
a salesman who must visit each city exactly once and
minimize the time of the complete tour. It is assumed that the
speed of the salesman is constant (vc) and he tries to minimize
the time of the tour**3. 
The objective function is given by:
f(x)=k=1∑n−1​txk​,xk+1​​+txn​,x1​​.

```python


# ---- init & operators ----
from extension.kp_repair.cheap_mutate_repair import MutateKPWithRepairAndStress


init_population_ttp_obj = InitPopulationHybrid(method="TTP_hybrid", dataset=dataset)

crossover_tsp_obj = CrossoverMateiTSP("mixt", p_select=[0.7, 0.3])

# base KP crossover
crossover_kp_obj = CrossoverMateiKP("mixt", p_select=[0.1, 0.1, 0.1, 0.3, 0.4])

mutate_tsp_obj = MutateMateiTSP(
    "mixt",
    dataset=dataset,
    subset_size=40,
    p_select=[0.4, 0.3, 0.3],
)


# ---- Multi-strategy KP repair ----
multi_repair = MultiStrategyKPRepair(
    w        = dataset["item_weight"],
    v        = dataset["item_profit"],
    Wmax     = 25936,
    distance = dataset["distance"],
    beta     = 1.0,
    p_modes  = (0.4, 0.4, 0.2),  # Fast, Horizon, aggressive-Fast
)

# ---- wrap KP operators ----
crossover_kp_obj = CrossoverKPWithRepair(
    base_crossover = crossover_kp_obj,
    repair         = multi_repair,   # multi-strategy here
    W              = 25936,
    dataset        = dataset,
)


mutate_kp_base = MutateMateiKP(
    method="mixt_extended",
    rate=0.035,      # slight increase
    p_select=[0.4, 0.3, 0.3],
)


mutate_kp_obj = MutateKPWithRepairAndStress(
    base_mutator=mutate_kp_base,
    W=25936,  # knapsack capacity
    profits=dataset["item_profit"],
    weights=dataset["item_weight"],
    stress_destroy_frac=0.35,
    stress_destroy_worst_frac=0.15,

    allow_add=True,
    repair=multi_repair,
)

fitness_obj = TTPFitness("TTP_standard", R=5.61, W=25936)
metrics_obj = MetricsTTPMATEI("TTP_linear", dataset, v_min=0.1, v_max=1, W=25936)

stres_obj = StresTTP(
    plateau_window=15,
    plateau_rtol=1e-3,
    replace_ratio=0.20,       # was 0.4 – too destructive
    mutation_boost=True,
    mutation_boost_factor=1.25,  # was 1.5
    mutation_rate_max=0.12,      # was 0.25+ explosive
    dynamic_alpha=True
)



# plug dependencies into stress
stres_obj.fitness   = fitness_obj
stres_obj.mutate_kp = mutate_kp_obj

elite_search_obj = TTPVNDLocalSearch(
    dataset=dataset,
    v_max=1.0, v_min=0.1, W=25936,
    R=5.61,
    max_rounds=5,    # a bit lighter LS, but more frequent via ELITE_FREQ=4
    use_kp_ls=True,
)

sel_parent1_obj = SelectParent("tour", size_subset=3)
sel_parent2_obj = SelectParent("tour_choice", size_subset=3)
```

```python

filename_logs = "{}/logs/history.csv".format(LCL_PATH)
extern_commnad_file = "{}/extern_command.cmd".format(EXTERN_CMD_PATH)

ttp = GeneticAlgorithmWithEliteSearch(
    name="test",
    extern_commnad_file=extern_commnad_file,
    manager = {
        "subset_size": 5,
        "update_elite_fitness": True,

        # Adaptive SELECT_RATE: many parents early → fewer (stronger) later
        "select_rate_max": 0.85,   # gen 0: ~85% of population can be parents
        "select_rate_min": 0.35,   # last gen: ~35% can be parents
        "select_rate_schedule": "linear",
    },

    genoms={
        "check_freq": 50,
        "tsp": (0, dataset["GENOME_LENGTH"]),
        "kp":  (0, dataset["item_profit"].shape[0]),
    },

    init_population=init_population_ttp_obj,
    metric=metrics_obj,

    elite_search=elite_search_obj,
    elite_chrom="",

    fitness=fitness_obj,
    select_parent1=sel_parent1_obj,
    select_parent2=sel_parent2_obj,

    crossover={
        "method": "chromosome",
        "tsp": crossover_tsp_obj,
        "kp":  crossover_kp_obj,
    },
    mutate={
        "method": "chromosome",
        "tsp": mutate_tsp_obj,
        "kp":  mutate_kp_obj,
    },
    stres=stres_obj,
    callback={"filename": filename_logs, "freq": 10},
)
```

```python
ttp.setParameters(
    GENOME_LENGTH   = dataset["GENOME_LENGTH"],  # 280

    # Keep total evals in the same ballpark but smoother evolution:
    GENERATIONS     = 1000,   # more, but each gen a bit “lighter”
    POPULATION_SIZE = 400,    # slightly smaller population

    # Global mutation / crossover:
    MUTATION_RATE   = 0.08,   # was 0.06
    CROSSOVER_RATE  = 0.70,   # was 0.90

    # Base SELECT_RATE (will be overwritten by manager’s schedule anyway):
    SELECT_RATE     = 0.5,

    dataset = dataset,
    W       = 25936,
    v_min   = 0.1,
    v_max   = 1.0,
    R       = 5.61,

    # Elites + local search frequency:
    ELITE_SIZE = 40,  # was 25, ~10% of population now
    ELITE_FREQ = 8,   # run LS every 4 generations instead of 5
)


best_individ = ttp()
```
