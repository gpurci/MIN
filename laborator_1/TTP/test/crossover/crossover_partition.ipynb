## Similarity test

```python

def similarity_gene(parent1, parent2):
    arr1, arr2 = np.nonzero(parent1 == parent2.reshape(-1, 1))
    #print("arr1 {},\narr2 {}".format(arr1, arr2))
    mask = np.abs(arr2[1:]-arr2[:-1])==1
    #print("mask", mask)
    #
    sim_map = []
    lenght = 0
    start_idx = 0
    for idx, bit in enumerate(mask):
        if (bit == False):
            sim_map.append([start_idx, start_idx+lenght+1])
            start_idx = idx+1
            lenght = 0
        else:
            lenght += 1
    else:
        sim_map.append([start_idx, start_idx+lenght+1])
    return sim_map

def show_similarity(sim_map):
    for start_idx, end_idx in sim_map:
        print(parent2[start_idx:end_idx])
```

```python

parent1 = np.arange(20)
parent2 = parent1.copy()
sim_map = similarity_gene(parent1, parent2)
show_similarity(sim_map)
```

```python

parent2 = parent1.copy()
parent2[7:13] = np.flip(parent2[7:13])
sim_map = similarity_gene(parent1, parent2)
show_similarity(sim_map)
```

```python

parent2 = parent1.copy()
parent2[7:13] = np.random.permutation(parent2[7:13])
sim_map = similarity_gene(parent1, parent2)
show_similarity(sim_map)
```

```python

parent2 = parent1.copy()
parent2[7:13] = np.roll(parent2[7:13], 2)
sim_map = similarity_gene(parent1, parent2)
show_similarity(sim_map)
```

```python

parent2 = parent1.copy()
parent2 = np.roll(parent2, 3)
parent2[7:13] = np.roll(parent2[7:13], 2)
sim_map = similarity_gene(parent1, parent2)
show_similarity(sim_map)
```

```python

parent2 = parent1.copy()
parent2[1:] = np.roll(parent2[1:], 3)
parent2[7:13] = np.flip(parent2[7:13])
sim_map = similarity_gene(parent1, parent2)
show_similarity(sim_map)
```

```python

parent2 = parent1.copy()
parent2[1:] = np.roll(parent2[1:], 3)
parent2[7:13] = np.flip(parent2[7:13])
parent2[13:] = np.random.permutation(parent2[13:])
print("parent2", parent2)
sim_map = similarity_gene(parent1, parent2)
show_similarity(sim_map)
```
